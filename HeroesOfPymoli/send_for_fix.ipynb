# Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0,9,14,19,24,29,34,39,40]
age_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35_39","40+"]
purchase_data["Age Group"]=pd.cut(purchase_data["Age"], age_bins, labels=age_labels)
age_group=purchase_data.groupby("Age Group")
grouped_age=age_group["SN"].nunique()
grouped_by_age=pd.DataFrame({"Total Count":grouped_age})
#del grouped_by_age.index.name
#calculating Percentage
grouped_by_age["Percentage of Players"]=(grouped_by_age["Total Count"]/total_players)*100
grouped_by_age["Percentage of Players"]=grouped_by_age["Percentage of Players"].map("{:.2f}".format)

grouped_by_age
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
retrive_df=purchase_data[["Item ID","Item Name","Price"]]
retrive_df.head()
grouped_iteam=purchase_data.groupby(['Item ID', 'Item Name'])
#grouped_iteam.count()
grouped_iteam_purchase_count=grouped_iteam["Price"].count()
grouped_iteam_purchase_count
grouped_iteam_purchase_price=(grouped_iteam["Price"].sum())
iteam_Price= grouped_iteam_purchase_price/grouped_iteam_purchase_count
#grouped_iteam.purchase_price
populer_items=pd.DataFrame({"Purchase Count":grouped_iteam_purchase_count,
                            "Iteam Price":iteam_Price,
                            "Total Purchase Value":grouped_iteam_purchase_price})
popular_items_ST=populer_items.sort_values(["Purchase Count"],ascending=False)
popular_items_ST.style.format({"Item price":"${:,.2f}",
                               "Total Purchase Value":"${:,.2f}"})


```
